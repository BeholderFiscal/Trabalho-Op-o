```python
import pandas as pd
from unidecode import unidecode

planilhagerencial = pd.read_excel('lista gerencial de conhecimentos.xls')

documentoteste = planilhagerencial[(planilhagerencial['Empresa'] == 3) & (planilhagerencial['Frete Pago'] == 'Remetente')]

documentoteste = documentoteste[['Remetente - CNPJ/CPF', 'Destinatário - CNPJ/CPF', 'Notas Fiscais']].dropna()

# Converte as colunas relevantes para números de ponto flutuante e, em seguida, realiza a soma
documentoteste['soma dos cnpj'] = (pd.to_numeric(documentoteste['Remetente - CNPJ/CPF'], errors='coerce') + pd.to_numeric(documentoteste['Destinatário - CNPJ/CPF'], errors='coerce')).astype(float)

# Converte a soma para inteiro, se desejar
documentoteste['soma dos cnpj'] = documentoteste['soma dos cnpj'].astype('Int64')

def calcular_soma(valor):
    if ',' in str(valor):
        return sum(map(int, str(valor).split(',')))
    else:
        return valor
documentoteste['Soma Separada'] = documentoteste['Notas Fiscais'].apply(calcular_soma)


documentoteste.to_excel('asadsa.xls')
display(documentoteste)
```
